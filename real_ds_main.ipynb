```python
artists_df = pd.read_csv('./raw_ds/hetrec2011-lastfm-2k/artists.dat',sep='\t')
# tags_df = pd.read_csv('./raw_ds/hetrec2011-lastfm-2k/tags.dat',sep='\t', encoding='latin-1')
user_artists_df = pd.read_csv('./raw_ds/hetrec2011-lastfm-2k/user_artists.dat',sep='\t')
# user_taggedartists_df = pd.read_csv('./raw_ds/hetrec2011-lastfm-2k/user_taggedartists.dat',sep='\t')
```

```python
unique_userID = np.unique(user_artists_df['userID'].values)
unique_artistID = np.unique(artists_df['id'].values)
p_dist = np.zeros((unique_userID.shape[0], unique_artistID.shape[0]))
print(p_dist.shape)
for _, row in user_artists_df.iterrows():
    u_idx = np.where(unique_userID==row['userID'])[0][0]
    a_idx = np.where(unique_artistID==row['artistID'])[0][0]
    p_dist[u_idx, a_idx] = row['weight']
# for i in range(p_dist.shape[0]):
#     p_dist[i] /= p_dist[i].sum()
```

```python
s = p_dist.sum(axis = 0)
print(s)
print(np.where(s<2))
```

```python
from collections import Counter
c = Counter(np.argmax(p_dist, axis=1))
# print(c.most_common())
count_threshold = 13
l = []
for tmp in c.most_common():
    artistID, count = tmp
    if count < count_threshold:
        continue
    l.append(artistID)
print(f"Keeping only {len(l)}/17632 artists that are most favorited by more than {count_threshold}/1892 people.")
```
